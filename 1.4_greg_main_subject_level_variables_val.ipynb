```python
import pandas as pd
import numpy as np
from os import path,listdir
```

```python
location = path.join('main_trial_level_cleaned.csv')
sheet=pd.read_csv(location) #opening the excel file - trial level data
#sheet.head()
sheet.tail()
```

```python
df = sheet.groupby(["SubjNum","outcome"]).mean().reset_index()  #groupig things by subjnum and stock value
df.head()
```

we made G10, G2, L10, L2 here:

```python
four_errors = df.pivot(index='SubjNum',columns='outcome',values='val_error_3sd_removed') #pivot it into what we want: each subject has four prob. errors
```

```python
#create a column called magnitude, which greg's data do not come with
sheet['Magnitude'] = 'LowMag'
for i in sheet.index:
    if sheet.at[i, 'outcome'] == (10 or -10):
        sheet.at[i, 'Magnitude'] = 'HighMag'
    elif sheet.at[i, 'outcome'] == (2 or -2):
        sheet.at[i, 'Magnitude'] = 'LowMag'
```

```python
mag_df = sheet.groupby(["SubjNum","Magnitude"]).mean().reset_index()  #groupig data by magnitude
mag_pivot = mag_df.pivot(index='SubjNum',columns='Magnitude',values='val_error_3sd_removed') #pivot it into what we want: each subject has high and low magnitude
mag_pivot
```

```python
domain_df = sheet.groupby(["SubjNum","attribcurrentcondition"]).mean().reset_index()  #groupig data by domain
domain_df.head()
domain_pivot = domain_df.pivot(index='SubjNum',columns='attribcurrentcondition',values='val_error_3sd_removed') #pivot it into what we want: each subject has high and low magnitude
domain_pivot
domain_df.head()
```

```python
result1=sheet.groupby('SubjNum', as_index=False).mean().merge(four_errors, on='SubjNum') #create a sheet with 4 errors
result2=result1.groupby('SubjNum', as_index=False).mean().merge(mag_pivot, on='SubjNum') #create a sheet with low/high magnitude
result3=result2.groupby('SubjNum', as_index=False).mean().merge(domain_pivot, on='SubjNum') #create a sheet with low/high magnitude
result3.head()
```

```python
#create columns called incongruence and domain_based_bias
result3['incongruence'] = result3['HighMag'] - result3['LowMag']
result3['domain_based_bias'] = result3['Gain'] - result3['Loss']
result3['stroop_ratio'] = result3['stroop_color_word']/result3['stroop_color']
```
