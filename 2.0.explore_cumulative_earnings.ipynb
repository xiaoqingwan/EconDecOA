```python
import pandas as pd
import numpy as np
from os import path,listdir

location = path.join('main_trial_level_cleaned.csv')
sheet = pd.read_csv(location)
# sheet = pd.read_csv('testing_four_bins.csv')
```

```python
###sort data into four bins based on the ordered value of total payout###
```

```python
sheet['bins'] = sheet.groupby('SubjNum').TotalPayout.apply(pd.cut, bins=4, labels = ["bin1","bin2","bin3","bin4"])
#binned all the trials into 4 bins based on subjective total payout
```

```python
df = sheet.groupby(["SubjNum","bins"]).mean().reset_index()  
#groupig things by subjnum and bins
df
```

```python
four_bins = df.pivot(index='SubjNum',columns='bins',values='val_error_3sd_removed') #pivot it into what we want: each subject has four prob. errors
```

```python
result=sheet.groupby('SubjNum', as_index=False).mean().merge(four_bins, on='SubjNum') #create a sheet with 4 errors
result.rename(columns={'SubjNum': 'sub'}, inplace=True)
result.tail()
```

```python
result.to_csv("subject_level_bins.csv")
```
