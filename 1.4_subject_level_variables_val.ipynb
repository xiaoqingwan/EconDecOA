```python
import pandas as pd
import numpy as np
from os import path,listdir
```

```python
location = path.join('..','derivatives', 'trialwise','main_trial_level_cleaned.csv')
nback_subject = pd.read_csv(path.join('..','derivatives', 'subjectwise','nback_subject_level_8.23.2019.csv'))
procspeed_df = pd.read_csv(path.join('..','derivatives', 'finaldata','procspd_subject_level_8.23.2019.csv'))
sheet=pd.read_csv(location) #opening the excel file - trial level data
nback_subject.head()
```

```python
df = sheet.groupby(["SubjNum","StockValue"]).mean().reset_index()  #groupig things by subjnum and stock value
df.head()
```

we made G10, G2, L10, L2 here:

```python
four_errors = df.pivot(index='SubjNum',columns='StockValue',values='val_error_3sd_removed') #pivot it into what we want: each subject has four prob. errors
```

```python
mag_df = sheet.groupby(["SubjNum","Magnitude"]).mean().reset_index()  #groupig data by magnitude
mag_df.head()
mag_pivot = mag_df.pivot(index='SubjNum',columns='Magnitude',values='val_error_3sd_removed') #pivot it into what we want: each subject has high and low magnitude
```

```python
domain_df = sheet.groupby(["SubjNum","Domain"]).mean().reset_index()  #groupig data by domain
domain_df.head()
domain_pivot = domain_df.pivot(index='SubjNum',columns='Domain',values='val_error_3sd_removed') #pivot it into what we want: each subject has high and low magnitude
domain_pivot
domain_df.head()
```

```python
result1=sheet.groupby('SubjNum', as_index=False).mean().merge(four_errors, on='SubjNum') #create a sheet with 4 errors
result2=result1.groupby('SubjNum', as_index=False).mean().merge(mag_pivot, on='SubjNum') #create a sheet with low/high magnitude
result3=result2.groupby('SubjNum', as_index=False).mean().merge(domain_pivot, on='SubjNum') #create a sheet with gain/loss domains
result3.head()
```

## Merge main task and nback and procspeed data

```python
result3.rename(columns={'SubjNum': 'sub'}, inplace=True)
result3.rename(columns={'high': 'HighMag'}, inplace=True)
result3.rename(columns={'low': 'LowMag'}, inplace=True)
procspeed_df.rename(columns={'Subject': 'sub'}, inplace=True)
procspeed_df.rename(columns={'RT': 'procspeed'}, inplace=True)
result3['incongruence'] = result3['HighMag'] - result3['LowMag']
result3['domain_based_bias'] = result3['GAIN'] - result3['LOSS']
main_task_nback = pd.merge(result3, nback_subject, how='left', on="sub")
main_task_nback_procspeed = pd.merge(main_task_nback, procspeed_df, how='outer', on="sub")
```

```python
derivs_dir = path.join('..','derivatives')
output_path = path.join(derivs_dir,'subjectwise',"main_subject_level_val_8.28.2019.csv")
main_task_nback_procspeed.to_csv(output_path) #this outputs in the code directory
```
