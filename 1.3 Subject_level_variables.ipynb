```python
import pandas as pd
import numpy as np
from os import path,listdir
```

```python
location = path.join('main_trial_level_cleaned.csv')
sheet=pd.read_csv(location) #opening the excel file - trial level data
#sheet.head()
sheet.tail()
```

```python
list(sheet.columns.values)
```

```python
df = sheet.groupby(["SubjNum","outcome"]).mean().reset_index()  #groupig things by subjnum and stock value
df.head()
```

we made G10, G2, L10, L2 here:

```python
four_errors = df.pivot(index='SubjNum',columns='outcome',values='abs_error_3sd_removed') #pivot it into what we want: each subject has four prob. errors
```

```python
result=sheet.groupby('SubjNum', as_index=False).mean().merge(four_errors, on='SubjNum') #create a sheet with 4 errors
result.tail()
```

## Merge main task and nback data

```python
result.rename(columns={'SubjNum': 'sub'}, inplace=True)
#merge nback and main task data. how='left' means keep the young adults who do not have nback data.
```
